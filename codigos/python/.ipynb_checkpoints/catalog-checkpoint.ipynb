```python
import json
import pandas as pd
```

```python
with open('catalog.json') as f:
    datas = json.load(f)
```

```python
df = pd.DataFrame(datas)
```

```python
new_df = df[['simulation_name','name','initial_mass_ratio','initial_dimensionless_spin1_mag','initial_dimensionless_spin2_mag']].copy()
mass_ratio = df['initial_mass_ratio'].copy()
for i in range(len(mass_ratio)):
    mass_ratio[i] = round(mass_ratio[i],0)
new_df['initial_mass_ratio'] = mass_ratio
```

```python
spin1 = df['initial_dimensionless_spin1_mag'].copy()
for i in range(len(spin1)):
    spin1[i] = round(spin1[i],8)
new_df['initial_dimensionless_spin1_mag'] = spin1
```

```python
spin2 = df['initial_dimensionless_spin2_mag'].copy()
for i in range(len(spin2)):
    spin2[i] = round(spin2[i],8)
new_df['initial_dimensionless_spin2_mag'] = spin2
```

```python
for name, group in new_df.groupby(['initial_dimensionless_spin1_mag','initial_dimensionless_spin2_mag']):
    print(name)
    print(group[['initial_mass_ratio']])
```
