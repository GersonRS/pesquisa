```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import lag_plot
from pandas.plotting import autocorrelation_plot
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
dados = pd.Series([200,202,208,204,204,207,207,204,202,199,201,198,200,202,203,205,207,211,204,206,203,203,201,198,200,206,207,206,200,203,203,200,200,195,202,204])
```

```python
plt.figure(figsize=(20,10))
dados.plot()
```

```python
plt.figure(1)
plt.figure(figsize=(10,5))
lag_plot(dados,lag=2)
plt.figure(2)
plt.figure(figsize=(10,5))
lag_plot(dados,lag=2)
```

```python
autocorrelation_plot(dados)
```

```python
plot_acf(dados)
```

```python
desvio = np.zeros(len(dados))
for i in range(len(dados)):
    desvio[i] = dados[i]-dados.mean()

C = np.zeros(3)
R = np.zeros(3)
K = [0,1,2]
N = len(dados)
for k in K:
    somatorio = 0
    for i in range(N-k):
        somatorio = somatorio + (desvio[i]*desvio[i+k])
    resultado = somatorio/N
    C[k] = resultado

for i in range(3):
    R[i] = C[i]/C[0]
print(C)
print(R)
```

```python
plt.bar(K,R,width=0.01)
```
